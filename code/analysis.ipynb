```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print("Imported libraries sucessfully....")
```

```python
#get the raw data
from openpyxl import load_workbook as loadGold

gold = loadGold("../raw data/gold.xlsx")
sheet = gold.active

print("Successfully Loaded file: ", sheet.title)
```

```python
#figuring out syntax
print(sheet["A4"].value)
```

```python
#counting total data points
#from January 2013 till September 2025
syear=2013
smonth=1
eyear=2025
emonth=9
c = (eyear-syear)*12 + emonth - smonth + 1
print(c)
```

```python
#extracting useful data
gvalues = []
count = 0
for row in sheet.iter_rows(min_row=8, values_only=True):
    v = row[2]
    n = row[1]
    if v is not None and n != "Average" and count < c:
        gvalues.append(v)
        gvalues[count] = round(gvalues[count], 2)
        count += 1
for g in gvalues:
    print(g)
```

```python
#dataframe syntax learning
g = pd.DataFrame(gvalues, columns = ["Gold Price (Rs.)"])
print(g)
```

```python
#making the data frame

dates = pd.date_range(start='2013-01-01', end='2025-09-01', freq='MS')

goldata = pd.DataFrame({
    'Dates': dates,
    'Gold Price': gvalues
})
print(goldata)
```
